```python
import matplotlib.pyplot as plt             
import cv2                                 
import tensorflow as tf  
from keras.models import load_model
import numpy as np
model = load_model('final_model.h5')
class_names = ['land', 'water']
```

### Feel free to take a screenshot from Google Earth and save it in a folder called "modeltest" and see if it predicts correctly!

```python
print("Enter image name:")
img_name = input()+'.PNG'
img = cv2.imread('modeltest/'+img_name)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.resize(img, dsize=(150,150), interpolation=cv2.INTER_CUBIC)
plt.imshow(res.astype(np.uint8))
prediction = model.predict(np.array([res]))     # Vector of probabilities
pred_label = np.argmax(prediction, axis = 1) # We take the highest probability
if pred_label == 0:
    print('\n')
    plt.title('Predicted = ' +class_names[pred_label[0]],color="red",fontsize=20)
else:
    print('\n')
    plt.title('Predicted = ' +class_names[pred_label[0]],color="green",fontsize=20)
```
